```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import imageio
import io
import pandas as pd
from tqdm import tqdm
from typing import Union, Optional, Dict, Tuple, List
import pickle
from sklearn.model_selection import train_test_split

import torch
from torch_geometric.data import Data
from mpl_toolkits.mplot3d import Axes3D

class TennisServeProcessor:
    """
    A unified class to visualize tennis serve data and convert it 
    to PyTorch Geometric datasets for next-step prediction.
    This version includes automatic player ID mapping, data normalization,
    and train/test/validation splits while keeping serves as separate files.
    """
    
    def __init__(self, base_dir: Union[str, Path]):
        self.base_dir = Path(base_dir)
        self.metadata_path = self.base_dir / "metadata" / "metadata.csv"
        self.joint_names = [
            "Pelvis", "R Hip", "R Knee", "R Ankle", "L Hip", "L Knee", "L Ankle", 
            "Spine", "Thorax", "Neck", "Head", "L Shoulder", "L Elbow", "L Wrist", 
            "R Shoulder", "R Elbow", "R Wrist"
        ]
        self.skeleton_edges = [
            (0, 1), (1, 2), (2, 3), (0, 4), (4, 5), (5, 6), (0, 7), 
            (7, 8), (8, 9), (9, 10), (8, 11), (11, 12), (12, 13), 
            (8, 14), (14, 15), (15, 16)
        ]
        self.metadata = self._load_metadata()
        
        # Create a unique numerical ID for each player
        self.servers = sorted(self.metadata['server'].unique())
        self.player_to_id: Dict[str, int] = {name: i for i, name in enumerate(self.servers)}
        print(f"Created mapping for {len(self.player_to_id)} unique players.")
        
        # Normalization statistics (will be computed during processing)
        self.normalization_stats = {}

    def _load_metadata(self) -> pd.DataFrame:
        if not self.metadata_path.exists():
            raise FileNotFoundError(f"Metadata not found at {self.metadata_path}")
        print("Metadata loaded.")
        return pd.read_csv(self.metadata_path)

    def read_serve_keypoints(self, serve_row: pd.Series) -> Optional[np.ndarray]:
        """Reads the keypoint data for a single serve."""
        keypoints_path = self.base_dir / serve_row['keypoints_path']
        if not keypoints_path.exists():
            return None
        with open(keypoints_path, 'r') as f:
            return np.array(json.load(f), dtype=np.float32)

    def compute_normalization_statistics(self) -> Dict[str, np.ndarray]:
        """
        Compute normalization statistics (mean, std) for positions and velocities
        across all serves in the dataset.
        """
        print("\nüìä Computing normalization statistics...")
        
        all_positions = []
        all_velocities = []
        
        for _, row in tqdm(self.metadata.iterrows(), total=len(self.metadata), desc="Computing stats"):
            keypoints = self.read_serve_keypoints(row)
            if keypoints is None or len(keypoints) < 2:
                continue
                
            positions = keypoints  # Shape: (T, 17, 3)
            velocities = np.diff(positions, axis=0, append=positions[-1:,:,:])
            
            all_positions.append(positions.reshape(-1, 3))
            all_velocities.append(velocities.reshape(-1, 3))
        
        if not all_positions:
            raise ValueError("No valid keypoint data found for computing statistics!")
        
        # Concatenate all data
        all_positions = np.vstack(all_positions)
        all_velocities = np.vstack(all_velocities)
        
        # Compute statistics
        stats = {
            'position_mean': np.mean(all_positions, axis=0),
            'position_std': np.std(all_positions, axis=0),
            'velocity_mean': np.mean(all_velocities, axis=0),
            'velocity_std': np.std(all_velocities, axis=0),
        }
        
        # Ensure std is not zero (add small epsilon)
        epsilon = 1e-8
        stats['position_std'] = np.maximum(stats['position_std'], epsilon)
        stats['velocity_std'] = np.maximum(stats['velocity_std'], epsilon)
        
        self.normalization_stats = stats
        print("‚úÖ Normalization statistics computed.")
        return stats

    def normalize_data(self, positions: np.ndarray, velocities: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
        """
        Normalize positions and velocities using computed statistics.
        
        Args:
            positions: Array of shape (T, 17, 3)
            velocities: Array of shape (T, 17, 3)
            
        Returns:
            Normalized positions and velocities
        """
        if not self.normalization_stats:
            raise ValueError("Normalization statistics not computed! Call compute_normalization_statistics() first.")
        
        original_shape = positions.shape
        
        # Reshape for normalization
        pos_flat = positions.reshape(-1, 3)
        vel_flat = velocities.reshape(-1, 3)
        
        # Normalize
        pos_normalized = (pos_flat - self.normalization_stats['position_mean']) / self.normalization_stats['position_std']
        vel_normalized = (vel_flat - self.normalization_stats['velocity_mean']) / self.normalization_stats['velocity_std']
        
        # Reshape back
        pos_normalized = pos_normalized.reshape(original_shape)
        vel_normalized = vel_normalized.reshape(original_shape)
        
        return pos_normalized, vel_normalized

    def denormalize_positions(self, normalized_positions: np.ndarray) -> np.ndarray:
        """
        Denormalize positions back to original scale.
        
        Args:
            normalized_positions: Normalized positions array
            
        Returns:
            Denormalized positions
        """
        if not self.normalization_stats:
            raise ValueError("Normalization statistics not available!")
        
        original_shape = normalized_positions.shape
        pos_flat = normalized_positions.reshape(-1, 3)
        
        denormalized = pos_flat * self.normalization_stats['position_std'] + self.normalization_stats['position_mean']
        
        return denormalized.reshape(original_shape)

    def create_data_splits(self, test_size: float = 0.2, val_size: float = 0.1, 
                          random_state: int = 42, stratify_by_player: bool = True) -> Dict[str, List[str]]:
        """
        Create train/test/validation splits for the serves.
        
        Args:
            test_size: Fraction of data for testing
            val_size: Fraction of data for validation (from remaining after test split)
            random_state: Random seed for reproducibility
            stratify_by_player: Whether to stratify splits by player to ensure balanced representation
            
        Returns:
            Dictionary with 'train', 'test', 'val' keys containing lists of serve_ids
        """
        print(f"\nüìä Creating data splits (test: {test_size:.1%}, val: {val_size:.1%})...")
        
        serve_ids = self.metadata['serve_id'].tolist()
        
        # Check player distribution for stratification
        player_counts = self.metadata['server'].value_counts()
        players_with_one_serve = (player_counts == 1).sum()
        total_players = len(player_counts)
        
        if stratify_by_player and players_with_one_serve > 0:
            print(f"‚ö†Ô∏è  Warning: {players_with_one_serve}/{total_players} players have only 1 serve.")
            print("   This can cause issues with stratified splitting.")
            
            # Check if stratification is feasible
            min_samples_needed = max(2, int(1 / min(test_size, val_size if val_size > 0 else 1)))
            players_sufficient = (player_counts >= min_samples_needed).sum()
            
            if players_with_one_serve > total_players * 0.3:  # If >30% of players have only 1 serve
                print("   Too many single-serve players. Switching to random splits.")
                stratify_by_player = False
            else:
                print("   Proceeding with stratification. Single-serve players will be handled carefully.")
        
        if stratify_by_player:
            try:
                servers = self.metadata['server'].tolist()
                # First split: train+val vs test
                train_val_ids, test_ids = train_test_split(
                    serve_ids, test_size=test_size, random_state=random_state,
                    stratify=servers
                )
                
                # Get corresponding servers for train_val set
                train_val_servers = [self.metadata[self.metadata['serve_id'] == sid]['server'].iloc[0] 
                                   for sid in train_val_ids]
                
                # Second split: train vs val
                if val_size > 0:
                    val_size_adjusted = val_size / (1 - test_size)  # Adjust val_size for remaining data
                    try:
                        train_ids, val_ids = train_test_split(
                            train_val_ids, test_size=val_size_adjusted, random_state=random_state,
                            stratify=train_val_servers
                        )
                    except ValueError as e:
                        print(f"   Warning: Stratification failed for train/val split: {e}")
                        print("   Using random split for train/val.")
                        train_ids, val_ids = train_test_split(
                            train_val_ids, test_size=val_size_adjusted, random_state=random_state
                        )
                else:
                    train_ids, val_ids = train_val_ids, []
                    
            except ValueError as e:
                print(f"   Warning: Stratification failed: {e}")
                print("   Falling back to random splits.")
                stratify_by_player = False
        
        if not stratify_by_player:
            # Random splits as fallback
            # First split: train+val vs test
            train_val_ids, test_ids = train_test_split(
                serve_ids, test_size=test_size, random_state=random_state
            )
            
            # Second split: train vs val
            if val_size > 0:
                val_size_adjusted = val_size / (1 - test_size)
                train_ids, val_ids = train_test_split(
                    train_val_ids, test_size=val_size_adjusted, random_state=random_state
                )
            else:
                train_ids, val_ids = train_val_ids, []
        
        splits = {
            'train': train_ids,
            'test': test_ids,
            'val': val_ids
        }
        
        # Report player distribution across splits
        if stratify_by_player:
            train_players = set(self.metadata[self.metadata['serve_id'].isin(train_ids)]['server'])
            test_players = set(self.metadata[self.metadata['serve_id'].isin(test_ids)]['server'])
            val_players = set(self.metadata[self.metadata['serve_id'].isin(val_ids)]['server']) if val_ids else set()
            
            print(f"‚úÖ Data splits created with stratification:")
            print(f"   Train: {len(train_ids)} serves ({len(train_ids)/len(serve_ids):.1%}) from {len(train_players)} players")
            print(f"   Test:  {len(test_ids)} serves ({len(test_ids)/len(serve_ids):.1%}) from {len(test_players)} players")
            if val_ids:
                print(f"   Val:   {len(val_ids)} serves ({len(val_ids)/len(serve_ids):.1%}) from {len(val_players)} players")
            
            # Check for player overlap (should be minimal with good stratification)
            overlap_train_test = len(train_players & test_players)
            overlap_train_val = len(train_players & val_players) if val_players else 0
            overlap_test_val = len(test_players & val_players) if val_players else 0
            
            if overlap_train_test + overlap_train_val + overlap_test_val > 0:
                print(f"   Player overlaps: Train-Test: {overlap_train_test}, Train-Val: {overlap_train_val}, Test-Val: {overlap_test_val}")
        else:
            print(f"‚úÖ Data splits created (random):")
            print(f"   Train: {len(train_ids)} serves ({len(train_ids)/len(serve_ids):.1%})")
            print(f"   Test:  {len(test_ids)} serves ({len(test_ids)/len(serve_ids):.1%})")
            if val_ids:
                print(f"   Val:   {len(val_ids)} serves ({len(val_ids)/len(serve_ids):.1%})")
        
        return splits

    def create_serve_animation_gif(self, serve_id: str, save_path: str = "serve_animation.gif", 
                                 fps: int = 15, use_normalized: bool = False):
        """Creates a 3D GIF animation of a single tennis serve."""
        serve_row = self.metadata[self.metadata['serve_id'] == serve_id].iloc[0]
        keypoints = self.read_serve_keypoints(serve_row)
        
        if keypoints is None:
            print(f"Could not generate animation for serve {serve_id}.")
            return

        # Optionally use normalized data for visualization
        if use_normalized and self.normalization_stats:
            velocities = np.diff(keypoints, axis=0, append=keypoints[-1:,:,:])
            keypoints, _ = self.normalize_data(keypoints, velocities)

        print(f"üé¨ Generating 3D animation for serve '{serve_id}'...")
        min_coords, max_coords = keypoints.min(axis=(0, 1)), keypoints.max(axis=(0, 1))
        center, max_range = (max_coords + min_coords) / 2, (max_coords - min_coords).max()
        
        with imageio.get_writer(save_path, mode='I', fps=fps, loop=0) as writer:
            for i in tqdm(range(keypoints.shape[0]), desc="Rendering frames"):
                fig = plt.figure(figsize=(8, 8))
                ax = fig.add_subplot(111, projection='3d')
                frame_kps = keypoints[i]
                ax.scatter(frame_kps[:, 0], frame_kps[:, 1], frame_kps[:, 2], c='red', s=40)
                for start, end in self.skeleton_edges:
                    ax.plot(*zip(frame_kps[start], frame_kps[end]), color='k')
                ax.set_xlim(center[0] - max_range/2, center[0] + max_range/2)
                ax.set_ylim(center[1] - max_range/2, center[1] + max_range/2)
                ax.set_zlim(center[2] - max_range/2, center[2] + max_range/2)
                ax.set_xlabel("X"); ax.set_ylabel("Y"); ax.set_zlabel("Z")
                title_suffix = " (Normalized)" if use_normalized else ""
                ax.set_title(f"Serve Animation{title_suffix} - Frame {i+1}/{keypoints.shape[0]}")
                ax.view_init(elev=20, azim=-60)
                
                with io.BytesIO() as buf:
                    plt.savefig(buf, format='png', dpi=90)
                    buf.seek(0)
                    writer.append_data(imageio.v2.imread(buf))
                plt.close(fig)
        print(f"‚úÖ Animation saved to '{save_path}'")

    def _create_graph_connectivity(self) -> torch.Tensor:
        """Helper to create the edge_index for the skeleton graph."""
        edges = []
        for i, j in self.skeleton_edges:
            edges.extend([[i, j], [j, i]])
        return torch.tensor(edges, dtype=torch.long).t().contiguous()

    def create_pyg_dataset_for_rollout(self, output_dir: Union[str, Path], 
                                     splits: Dict[str, List[str]], 
                                     include_velocities: bool = True,
                                     normalize_data: bool = True):
        """
        Processes all serves, saving each as a PyG sequence with optional normalization
        and organized by train/test/val splits.
        Modified to include 6 features (pos + vel) in both inputs AND targets.
        """
        output_dir = Path(output_dir)
        output_dir.mkdir(parents=True, exist_ok=True)
        
        # Create subdirectories for each split
        for split_name in splits.keys():
            (output_dir / split_name).mkdir(parents=True, exist_ok=True)
        
        edge_index = self._create_graph_connectivity()
        
        # Compute normalization statistics if needed
        if normalize_data:
            self.compute_normalization_statistics()
            
            # Save normalization statistics
            with open(output_dir / 'normalization_stats.pkl', 'wb') as f:
                pickle.dump(self.normalization_stats, f)
            print("‚úÖ Normalization statistics saved.")
        
        print(f"\nüß† Processing {len(self.metadata)} serves into PyG format...")
        
        processed_counts = {split: 0 for split in splits.keys()}
        
        for _, row in tqdm(self.metadata.iterrows(), total=len(self.metadata), desc="Creating PyG Serves"):
            keypoints = self.read_serve_keypoints(row)
            if keypoints is None or len(keypoints) < 2:
                continue

            serve_id = row['serve_id']
            
            # Determine which split this serve belongs to
            split_name = None
            for split, serve_list in splits.items():
                if serve_id in serve_list:
                    split_name = split
                    break
            
            if split_name is None:
                print(f"Warning: Serve {serve_id} not found in any split!")
                continue

            positions = keypoints
            velocities = np.diff(positions, axis=0, append=positions[-1:,:,:])
            
            # Apply normalization if requested
            if normalize_data:
                positions_norm, velocities_norm = self.normalize_data(positions, velocities)
                # For 6-feature model: concatenate positions and velocities
                node_features = np.concatenate([positions_norm, velocities_norm], axis=-1)
                target_features = np.concatenate([positions_norm, velocities_norm], axis=-1)
            else:
                # For 6-feature model: concatenate positions and velocities
                node_features = np.concatenate([positions, velocities], axis=-1)
                target_features = np.concatenate([positions, velocities], axis=-1)
            
            server_name = row['server']
            server_id = self.player_to_id.get(server_name, -1)

            serve_sequence = []
            for t in range(node_features.shape[0] - 1):
                x_t = torch.tensor(node_features[t], dtype=torch.float32)  # 6 features: pos + vel at time t
                pos_t = torch.tensor(positions[t] if not normalize_data else positions_norm[t], dtype=torch.float32)
                y_t = torch.tensor(target_features[t + 1], dtype=torch.float32)  # 6 features: pos + vel at time t+1
                
                data_t = Data(
                    x=x_t,  # Input: (17, 6) - positions and velocities at time t
                    y=y_t,  # Target: (17, 6) - positions and velocities at time t+1
                    pos=pos_t,  # Just positions for visualization/reference
                    edge_index=edge_index,
                    
                    # Global context features
                    server_id=torch.tensor([server_id], dtype=torch.long),
                    serve_number=torch.tensor([row['ServeNumber']], dtype=torch.float32),
                    set_number=torch.tensor([row['SetNo']], dtype=torch.float32),
                    game_number=torch.tensor([row['GameNo']], dtype=torch.float32),
                    point_number=torch.tensor([row['PointNumber']], dtype=torch.float32)
                )
                serve_sequence.append(data_t)
            
            if serve_sequence:
                save_path = output_dir / split_name / f"{serve_id}.pt"
                torch.save(serve_sequence, save_path)
                processed_counts[split_name] += 1
                
        print(f"‚úÖ PyG dataset creation complete:")
        for split_name, count in processed_counts.items():
            print(f"   {split_name}: {count} serves saved in '{output_dir / split_name}'")
        
        # Save metadata and splits info
        with open(output_dir / 'player_mapping.json', 'w') as f:
            json.dump(self.player_to_id, f, indent=2)
        
        with open(output_dir / 'data_splits.json', 'w') as f:
            json.dump(splits, f, indent=2)
            
        # Save dataset info - UPDATED for 6 features
        dataset_info = {
            'total_serves': len(self.metadata),
            'processed_serves': sum(processed_counts.values()),
            'splits': {split: len(serve_list) for split, serve_list in splits.items()},
            'normalization_applied': normalize_data,
            'velocities_included': True,  # Always True for 6-feature model
            'joint_names': self.joint_names,
            'num_joints': len(self.joint_names),
            'feature_dim': 6,  # Always 6 features: 3 pos + 3 vel
            'input_features': 6,  # x has 6 features per joint
            'target_features': 6,  # y has 6 features per joint
            'data_format': 'positions_and_velocities_concatenated'
        }
        
        with open(output_dir / 'dataset_info.json', 'w') as f:
            json.dump(dataset_info, f, indent=2)
            
        print("‚úÖ Additional metadata saved:")
        print("   - player_mapping.json: Player name to ID mapping")
        print("   - data_splits.json: Train/test/val split information")
        print("   - dataset_info.json: Dataset statistics and configuration")
        print(f"   - Data format: 6 features per joint (3 pos + 3 vel) in both x and y")
        if normalize_data:
            print("   - normalization_stats.pkl: Statistics for denormalization")


# ==================== MAIN WORKFLOW ====================

if __name__ == '__main__':
    BASE_DIR = Path("/home/jtb3sud/G-PARC/demos/tennis")
    PYG_OUTPUT_DIR = Path("/project/vil_baek/psaap/tennis/seq_tennis_data_normalized")

    processor = TennisServeProcessor(BASE_DIR)
    
    # --- Workflow Step 1: Create train/test/val splits ---
    splits = processor.create_data_splits(
        test_size=0.2,
        val_size=0.1,
        random_state=42,
        stratify_by_player=True
    )
    
    # --- Workflow Step 2: Visualize a sample serve (Optional) ---
    sample_serve_id = processor.metadata['serve_id'].iloc[0]
    
    # First create animation with original data
    processor.create_serve_animation_gif(
        serve_id=sample_serve_id, 
        save_path=f"{sample_serve_id}_original.gif"
    )
    
    # --- Workflow Step 3: Process all serves into PyG format with normalization ---
    processor.create_pyg_dataset_for_rollout(
        output_dir=PYG_OUTPUT_DIR,
        splits=splits,
        include_velocities=True,
        normalize_data=True
    )
    
    # --- Workflow Step 4: Create animation with normalized data for comparison ---
    processor.create_serve_animation_gif(
        serve_id=sample_serve_id,
        save_path=f"{sample_serve_id}_normalized.gif",
        use_normalized=True
    )
    
    print(f"\nüéâ Enhanced workflow complete! Your data in '{PYG_OUTPUT_DIR}' is now ready.")
    print("üìÅ Directory structure:")
    print("   ‚îú‚îÄ‚îÄ train/          # Training serve files (.pt)")
    print("   ‚îú‚îÄ‚îÄ test/           # Test serve files (.pt)")
    print("   ‚îú‚îÄ‚îÄ val/            # Validation serve files (.pt)")
    print("   ‚îú‚îÄ‚îÄ player_mapping.json")
    print("   ‚îú‚îÄ‚îÄ data_splits.json")
    print("   ‚îú‚îÄ‚îÄ dataset_info.json")
    print("   ‚îî‚îÄ‚îÄ normalization_stats.pkl")
    print("\nüöÄ Ready for training with proper normalization and data splits!")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
from tqdm import tqdm
from typing import Union, Optional, Dict, List
import pickle
import torch
from torch_geometric.data import Data
from mpl_toolkits.mplot3d import Axes3D
import imageio
import io

class TennisDataVisualizer:
    """
    A visualization tool for already-processed tennis serve data.
    Works with PyTorch Geometric datasets that have been split and normalized.
    """
    
    def __init__(self, processed_data_dir: Union[str, Path], original_data_dir: Union[str, Path] = None):
        """
        Initialize the visualizer.
        
        Args:
            processed_data_dir: Path to the processed PyG data directory
            original_data_dir: Path to original data (optional, for comparison visualizations)
        """
        self.processed_data_dir = Path(processed_data_dir)
        self.original_data_dir = Path(original_data_dir) if original_data_dir else None
        
        # Load metadata
        self._load_processed_metadata()
        
        # Define joint names and skeleton
        self.joint_names = [
            "Pelvis", "R Hip", "R Knee", "R Ankle", "L Hip", "L Knee", "L Ankle", 
            "Spine", "Thorax", "Neck", "Head", "L Shoulder", "L Elbow", "L Wrist", 
            "R Shoulder", "R Elbow", "R Wrist"
        ]
        self.skeleton_edges = [
            (0, 1), (1, 2), (2, 3), (0, 4), (4, 5), (5, 6), (0, 7), 
            (7, 8), (8, 9), (9, 10), (8, 11), (11, 12), (12, 13), 
            (8, 14), (14, 15), (15, 16)
        ]
        
    def _load_processed_metadata(self):
        """Load all the metadata from the processed dataset."""
        # Load dataset info
        with open(self.processed_data_dir / 'dataset_info.json', 'r') as f:
            self.dataset_info = json.load(f)
        
        # Load data splits
        with open(self.processed_data_dir / 'data_splits.json', 'r') as f:
            self.data_splits = json.load(f)
        
        # Load player mapping
        with open(self.processed_data_dir / 'player_mapping.json', 'r') as f:
            self.player_mapping = json.load(f)
            self.id_to_player = {v: k for k, v in self.player_mapping.items()}
        
        # Load normalization stats if available
        norm_stats_path = self.processed_data_dir / 'normalization_stats.pkl'
        if norm_stats_path.exists():
            with open(norm_stats_path, 'rb') as f:
                self.normalization_stats = pickle.load(f)
        else:
            self.normalization_stats = None
            
        print(f"‚úÖ Loaded metadata for {self.dataset_info['total_serves']} serves")
        print(f"   Splits: Train={len(self.data_splits['train'])}, Test={len(self.data_splits['test'])}, Val={len(self.data_splits['val'])}")
        print(f"   Normalization: {'Applied' if self.normalization_stats else 'Not applied'}")

    def get_available_serves(self, split: str = 'train') -> List[str]:
        """Get list of available serve IDs for a given split."""
        serve_files = list((self.processed_data_dir / split).glob('*.pt'))
        return [f.stem for f in serve_files]

    def load_processed_serve(self, serve_id: str, split: str = None) -> List[Data]:
        """
        Load a processed serve from the PyG dataset.
        
        Args:
            serve_id: The serve ID to load
            split: Which split to load from (train/test/val). If None, searches all splits.
        
        Returns:
            List of PyG Data objects representing the serve sequence
        """
        if split:
            serve_path = self.processed_data_dir / split / f"{serve_id}.pt"
            if serve_path.exists():
                return torch.load(serve_path, weights_only=False)
        else:
            # Search all splits
            for split_name in ['train', 'test', 'val']:
                serve_path = self.processed_data_dir / split_name / f"{serve_id}.pt"
                if serve_path.exists():
                    return torch.load(serve_path, weights_only=False)
        
        raise FileNotFoundError(f"Serve {serve_id} not found in any split")

    def load_original_serve(self, serve_id: str) -> Optional[np.ndarray]:
        """Load original serve data if original_data_dir is provided."""
        if not self.original_data_dir:
            print("‚ùå Original data directory not provided")
            return None
        
        # This assumes your original data structure - adjust as needed
        metadata_path = self.original_data_dir / "metadata" / "metadata.csv"
        if not metadata_path.exists():
            print("‚ùå Original metadata not found")
            return None
        
        metadata = pd.read_csv(metadata_path)
        serve_row = metadata[metadata['serve_id'] == serve_id]
        if serve_row.empty:
            print(f"‚ùå Serve {serve_id} not found in original metadata")
            return None
        
        keypoints_path = self.original_data_dir / serve_row.iloc[0]['keypoints_path']
        if not keypoints_path.exists():
            print(f"‚ùå Keypoints file not found: {keypoints_path}")
            return None
        
        with open(keypoints_path, 'r') as f:
            return np.array(json.load(f), dtype=np.float32)

    def denormalize_positions(self, normalized_positions: np.ndarray) -> np.ndarray:
        """Denormalize positions back to original scale."""
        if not self.normalization_stats:
            print("‚ùå No normalization statistics available!")
            return normalized_positions
        
        original_shape = normalized_positions.shape
        pos_flat = normalized_positions.reshape(-1, 3)
        
        denormalized = pos_flat * self.normalization_stats['position_std'] + self.normalization_stats['position_mean']
        
        return denormalized.reshape(original_shape)

    def extract_positions_from_serve(self, serve_sequence: List[Data], denormalize: bool = True) -> np.ndarray:
        """
        Extract position data from a PyG serve sequence.
        
        Args:
            serve_sequence: List of PyG Data objects
            denormalize: Whether to denormalize the positions
        
        Returns:
            Array of shape (T, 17, 3) containing positions
        """
        positions = []
        for data in serve_sequence:
            pos = data.pos.numpy()  # Shape: (17, 3)
            positions.append(pos)
        
        positions = np.array(positions)  # Shape: (T, 17, 3)
        
        if denormalize and self.normalization_stats:
            positions = self.denormalize_positions(positions)
        
        return positions

    def visualize_normalization_statistics(self, save_path: str = "normalization_stats.png"):
        """Visualize the normalization statistics."""
        if not self.normalization_stats:
            print("‚ùå No normalization statistics available!")
            return
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        pos_mean = self.normalization_stats['position_mean']
        pos_std = self.normalization_stats['position_std']
        vel_mean = self.normalization_stats['velocity_mean']
        vel_std = self.normalization_stats['velocity_std']
        
        coords = ['X', 'Y', 'Z']
        x_pos = np.arange(len(coords))
        
        # Position mean
        axes[0,0].bar(x_pos, pos_mean, color=['red', 'green', 'blue'], alpha=0.7)
        axes[0,0].set_title('Position Mean by Coordinate', fontsize=12, fontweight='bold')
        axes[0,0].set_xlabel('Coordinate')
        axes[0,0].set_ylabel('Mean Value')
        axes[0,0].set_xticks(x_pos)
        axes[0,0].set_xticklabels(coords)
        axes[0,0].grid(True, alpha=0.3)
        
        # Position std
        axes[0,1].bar(x_pos, pos_std, color=['red', 'green', 'blue'], alpha=0.7)
        axes[0,1].set_title('Position Standard Deviation by Coordinate', fontsize=12, fontweight='bold')
        axes[0,1].set_xlabel('Coordinate')
        axes[0,1].set_ylabel('Standard Deviation')
        axes[0,1].set_xticks(x_pos)
        axes[0,1].set_xticklabels(coords)
        axes[0,1].grid(True, alpha=0.3)
        
        # Velocity mean
        axes[1,0].bar(x_pos, vel_mean, color=['darkred', 'darkgreen', 'darkblue'], alpha=0.7)
        axes[1,0].set_title('Velocity Mean by Coordinate', fontsize=12, fontweight='bold')
        axes[1,0].set_xlabel('Coordinate')
        axes[1,0].set_ylabel('Mean Value')
        axes[1,0].set_xticks(x_pos)
        axes[1,0].set_xticklabels(coords)
        axes[1,0].grid(True, alpha=0.3)
        
        # Velocity std
        axes[1,1].bar(x_pos, vel_std, color=['darkred', 'darkgreen', 'darkblue'], alpha=0.7)
        axes[1,1].set_title('Velocity Standard Deviation by Coordinate', fontsize=12, fontweight='bold')
        axes[1,1].set_xlabel('Coordinate')
        axes[1,1].set_ylabel('Standard Deviation')
        axes[1,1].set_xticks(x_pos)
        axes[1,1].set_xticklabels(coords)
        axes[1,1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.show()
        print(f"‚úÖ Normalization statistics saved to '{save_path}'")

    def compare_normalized_vs_denormalized(self, serve_id: str, save_path: str = None):
        """Compare normalized vs denormalized serve data."""
        if save_path is None:
            save_path = f"{serve_id}_norm_vs_denorm_comparison.png"
        
        # Load processed serve
        serve_sequence = self.load_processed_serve(serve_id)
        
        # Extract positions (normalized and denormalized)
        positions_norm = self.extract_positions_from_serve(serve_sequence, denormalize=False)
        positions_denorm = self.extract_positions_from_serve(serve_sequence, denormalize=True)
        
        fig = plt.figure(figsize=(16, 8))
        
        # Normalized data plot
        ax1 = fig.add_subplot(121, projection='3d')
        frame_indices = np.linspace(0, len(positions_norm)-1, min(5, len(positions_norm)), dtype=int)
        colors = plt.cm.viridis(np.linspace(0, 1, len(frame_indices)))
        
        for i, (frame_idx, color) in enumerate(zip(frame_indices, colors)):
            frame_pos = positions_norm[frame_idx]
            ax1.scatter(frame_pos[:, 0], frame_pos[:, 1], frame_pos[:, 2], 
                       c=[color], s=30, alpha=0.7, label=f'Frame {frame_idx}')
            
            if i == 0 or i == len(frame_indices) - 1:
                for start, end in self.skeleton_edges:
                    ax1.plot([frame_pos[start, 0], frame_pos[end, 0]],
                            [frame_pos[start, 1], frame_pos[end, 1]],
                            [frame_pos[start, 2], frame_pos[end, 2]], 
                            color='gray', alpha=0.3, linewidth=0.5)
        
        ax1.set_xlabel('Normalized X')
        ax1.set_ylabel('Normalized Y')
        ax1.set_zlabel('Normalized Z')
        ax1.set_title('Normalized Serve Data', fontweight='bold')
        ax1.legend()
        
        # Denormalized data plot
        ax2 = fig.add_subplot(122, projection='3d')
        
        for i, (frame_idx, color) in enumerate(zip(frame_indices, colors)):
            frame_pos = positions_denorm[frame_idx]
            ax2.scatter(frame_pos[:, 0], frame_pos[:, 1], frame_pos[:, 2], 
                       c=[color], s=30, alpha=0.7, label=f'Frame {frame_idx}')
            
            if i == 0 or i == len(frame_indices) - 1:
                for start, end in self.skeleton_edges:
                    ax2.plot([frame_pos[start, 0], frame_pos[end, 0]],
                            [frame_pos[start, 1], frame_pos[end, 1]],
                            [frame_pos[start, 2], frame_pos[end, 2]], 
                            color='gray', alpha=0.3, linewidth=0.5)
        
        ax2.set_xlabel('X')
        ax2.set_ylabel('Y')
        ax2.set_zlabel('Z')
        ax2.set_title('Denormalized Serve Data', fontweight='bold')
        ax2.legend()
        
        plt.suptitle(f'Normalized vs Denormalized: {serve_id}', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.show()
        print(f"‚úÖ Comparison saved to '{save_path}'")

    def visualize_serve_features(self, serve_id: str, save_path: str = None):
        """Visualize all features (positions + velocities) for a serve."""
        if save_path is None:
            save_path = f"{serve_id}_features.png"
        
        serve_sequence = self.load_processed_serve(serve_id)
        
        # Extract features (positions + velocities if included)
        features = []
        for data in serve_sequence:
            features.append(data.x.numpy())  # Shape: (17, feature_dim)
        
        features = np.array(features)  # Shape: (T, 17, feature_dim)
        
        feature_dim = features.shape[-1]
        is_pos_vel = feature_dim == 6  # positions (3) + velocities (3)
        
        if is_pos_vel:
            positions = features[:, :, :3]
            velocities = features[:, :, 3:]
            
            fig, axes = plt.subplots(2, 3, figsize=(18, 10))
            
            coords = ['X', 'Y', 'Z']
            colors = ['red', 'green', 'blue']
            
            for i, (coord, color) in enumerate(zip(coords, colors)):
                # Position distributions
                pos_data = positions[:, :, i].flatten()
                axes[0, i].hist(pos_data, bins=50, alpha=0.7, color=color)
                axes[0, i].set_title(f'Position {coord} Distribution', fontweight='bold')
                axes[0, i].set_xlabel(f'{coord} Value')
                axes[0, i].set_ylabel('Frequency')
                axes[0, i].grid(True, alpha=0.3)
                
                # Velocity distributions
                vel_data = velocities[:, :, i].flatten()
                axes[1, i].hist(vel_data, bins=50, alpha=0.7, color=color)
                axes[1, i].set_title(f'Velocity {coord} Distribution', fontweight='bold')
                axes[1, i].set_xlabel(f'{coord} Velocity')
                axes[1, i].set_ylabel('Frequency')
                axes[1, i].grid(True, alpha=0.3)
                
                # Add statistics
                pos_mean, pos_std = np.mean(pos_data), np.std(pos_data)
                vel_mean, vel_std = np.mean(vel_data), np.std(vel_data)
                
                axes[0, i].text(0.02, 0.98, f'Œº={pos_mean:.2f}\nœÉ={pos_std:.2f}', 
                              transform=axes[0, i].transAxes, verticalalignment='top',
                              bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
                
                axes[1, i].text(0.02, 0.98, f'Œº={vel_mean:.2f}\nœÉ={vel_std:.2f}', 
                              transform=axes[1, i].transAxes, verticalalignment='top',
                              bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
            
            plt.suptitle(f'Feature Distributions for Serve: {serve_id}', fontsize=16, fontweight='bold')
        else:
            # Just positions
            fig, axes = plt.subplots(1, 3, figsize=(15, 5))
            coords = ['X', 'Y', 'Z']
            colors = ['red', 'green', 'blue']
            
            for i, (coord, color) in enumerate(zip(coords, colors)):
                pos_data = features[:, :, i].flatten()
                axes[i].hist(pos_data, bins=50, alpha=0.7, color=color)
                axes[i].set_title(f'Position {coord} Distribution', fontweight='bold')
                axes[i].set_xlabel(f'{coord} Value')
                axes[i].set_ylabel('Frequency')
                axes[i].grid(True, alpha=0.3)
                
                pos_mean, pos_std = np.mean(pos_data), np.std(pos_data)
                axes[i].text(0.02, 0.98, f'Œº={pos_mean:.2f}\nœÉ={pos_std:.2f}', 
                           transform=axes[i].transAxes, verticalalignment='top',
                           bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
            
            plt.suptitle(f'Position Distributions for Serve: {serve_id}', fontsize=16, fontweight='bold')
        
        plt.tight_layout()
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.show()
        print(f"‚úÖ Feature visualization saved to '{save_path}'")

    def visualize_joint_trajectories(self, serve_id: str, joint_indices: List[int] = None, 
                                   save_path: str = None):
        """Visualize trajectories of specific joints over time."""
        if save_path is None:
            save_path = f"{serve_id}_joint_trajectories.png"
        
        if joint_indices is None:
            joint_indices = [16, 15, 14]  # R Wrist, R Elbow, R Shoulder
        
        serve_sequence = self.load_processed_serve(serve_id)
        positions = self.extract_positions_from_serve(serve_sequence, denormalize=True)
        
        fig, axes = plt.subplots(len(joint_indices), 1, figsize=(12, 4*len(joint_indices)))
        if len(joint_indices) == 1:
            axes = [axes]
        
        time_steps = np.arange(len(positions))
        coords = ['X', 'Y', 'Z']
        colors = ['red', 'green', 'blue']
        
        for i, joint_idx in enumerate(joint_indices):
            joint_name = self.joint_names[joint_idx]
            
            for coord_idx, (coord, color) in enumerate(zip(coords, colors)):
                axes[i].plot(time_steps, positions[:, joint_idx, coord_idx], 
                           color=color, label=f'{coord}', linewidth=2)
            
            axes[i].set_title(f'{joint_name} Trajectory', fontweight='bold')
            axes[i].set_xlabel('Time Step')
            axes[i].set_ylabel('Position')
            axes[i].legend()
            axes[i].grid(True, alpha=0.3)
        
        plt.suptitle(f'Joint Trajectories for Serve: {serve_id}', fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.show()
        print(f"‚úÖ Joint trajectories saved to '{save_path}'")

    def create_serve_animation(self, serve_id: str, save_path: str = None, 
                             fps: int = 15, use_denormalized: bool = True):
        """Create an animation of a processed serve."""
        if save_path is None:
            save_path = f"{serve_id}_animation.gif"
        
        serve_sequence = self.load_processed_serve(serve_id)
        positions = self.extract_positions_from_serve(serve_sequence, denormalize=use_denormalized)
        
        print(f"üé¨ Generating animation for serve '{serve_id}'...")
        min_coords, max_coords = positions.min(axis=(0, 1)), positions.max(axis=(0, 1))
        center, max_range = (max_coords + min_coords) / 2, (max_coords - min_coords).max()
        
        with imageio.get_writer(save_path, mode='I', fps=fps, loop=0) as writer:
            for i in tqdm(range(len(positions)), desc="Rendering frames"):
                fig = plt.figure(figsize=(8, 8))
                ax = fig.add_subplot(111, projection='3d')
                frame_pos = positions[i]
                
                ax.scatter(frame_pos[:, 0], frame_pos[:, 1], frame_pos[:, 2], c='red', s=40)
                for start, end in self.skeleton_edges:
                    ax.plot(*zip(frame_pos[start], frame_pos[end]), color='k')
                
                ax.set_xlim(center[0] - max_range/2, center[0] + max_range/2)
                ax.set_ylim(center[1] - max_range/2, center[1] + max_range/2)
                ax.set_zlim(center[2] - max_range/2, center[2] + max_range/2)
                ax.set_xlabel("X"); ax.set_ylabel("Y"); ax.set_zlabel("Z")
                
                title_suffix = " (Denormalized)" if use_denormalized else " (Normalized)"
                ax.set_title(f"Serve{title_suffix} - Frame {i+1}/{len(positions)}")
                ax.view_init(elev=20, azim=-60)
                
                with io.BytesIO() as buf:
                    plt.savefig(buf, format='png', dpi=90)
                    buf.seek(0)
                    writer.append_data(imageio.v2.imread(buf))
                plt.close(fig)
        
        print(f"‚úÖ Animation saved to '{save_path}'")

    def analyze_dataset_statistics(self, save_path: str = "dataset_analysis.png"):
        """Analyze and visualize dataset-wide statistics."""
        print("üìä Analyzing dataset statistics...")
        
        # Collect statistics from all serves
        split_stats = {}
        
        for split_name in ['train', 'test', 'val']:
            serve_files = list((self.processed_data_dir / split_name).glob('*.pt'))
            
            sequence_lengths = []
            player_counts = {}
            
            for serve_file in tqdm(serve_files, desc=f"Analyzing {split_name}"):
                try:
                    serve_sequence = torch.load(serve_file, weights_only=False)
                    sequence_lengths.append(len(serve_sequence))
                    
                    # Get player info from first data point
                    if len(serve_sequence) > 0:
                        player_id = serve_sequence[0].server_id.item()
                        player_name = self.id_to_player.get(player_id, f"Unknown_{player_id}")
                        player_counts[player_name] = player_counts.get(player_name, 0) + 1
                        
                except Exception as e:
                    print(f"Error loading {serve_file}: {e}")
                    continue
            
            split_stats[split_name] = {
                'sequence_lengths': sequence_lengths,
                'player_counts': player_counts,
                'total_serves': len(sequence_lengths)
            }
        
        # Create visualizations
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        
        # Sequence length distributions
        for i, split_name in enumerate(['train', 'test', 'val']):
            if split_stats[split_name]['sequence_lengths']:
                axes[0, i].hist(split_stats[split_name]['sequence_lengths'], bins=30, alpha=0.7)
                axes[0, i].set_title(f'{split_name.title()} - Sequence Lengths')
                axes[0, i].set_xlabel('Number of Frames')
                axes[0, i].set_ylabel('Frequency')
                axes[0, i].grid(True, alpha=0.3)
                
                mean_len = np.mean(split_stats[split_name]['sequence_lengths'])
                axes[0, i].axvline(mean_len, color='red', linestyle='--', 
                                 label=f'Mean: {mean_len:.1f}')
                axes[0, i].legend()
        
        # Player distributions (top 10 players per split)
        for i, split_name in enumerate(['train', 'test', 'val']):
            player_counts = split_stats[split_name]['player_counts']
            if player_counts:
                # Get top 10 players
                top_players = sorted(player_counts.items(), key=lambda x: x[1], reverse=True)[:10]
                players, counts = zip(*top_players)
                
                axes[1, i].bar(range(len(players)), counts, alpha=0.7)
                axes[1, i].set_title(f'{split_name.title()} - Top 10 Players')
                axes[1, i].set_xlabel('Players')
                axes[1, i].set_ylabel('Number of Serves')
                axes[1, i].set_xticks(range(len(players)))
                axes[1, i].set_xticklabels([p[:10] for p in players], rotation=45, ha='right')
                axes[1, i].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        plt.show()
        
        # Print summary statistics
        print("\nüìà Dataset Summary:")
        for split_name, stats in split_stats.items():
            if stats['sequence_lengths']:
                print(f"\n{split_name.upper()}:")
                print(f"  Total serves: {stats['total_serves']}")
                print(f"  Avg sequence length: {np.mean(stats['sequence_lengths']):.1f}")
                print(f"  Min/Max sequence length: {np.min(stats['sequence_lengths'])}/{np.max(stats['sequence_lengths'])}")
                print(f"  Unique players: {len(stats['player_counts'])}")
        
        print(f"‚úÖ Dataset analysis saved to '{save_path}'")


# ==================== EXAMPLE USAGE ====================

if __name__ == '__main__':
    # Initialize the visualizer
    PROCESSED_DATA_DIR = Path("/project/vil_baek/psaap/tennis/seq_tennis_data_normalized")
    ORIGINAL_DATA_DIR = Path("/home/jtb3sud/G-PARC/demos/tennis")  # Optional
    
    visualizer = TennisDataVisualizer(
        processed_data_dir=PROCESSED_DATA_DIR,
        original_data_dir=ORIGINAL_DATA_DIR
    )
    
    # Get a sample serve for visualization
    available_serves = visualizer.get_available_serves('train')
    sample_serve_id = available_serves[0]
    print(f"\nüéæ Analyzing sample serve: {sample_serve_id}")
    
    # 1. Visualize normalization statistics
    visualizer.visualize_normalization_statistics("normalization_overview.png")
    
    # 2. Compare normalized vs denormalized data
    visualizer.compare_normalized_vs_denormalized(sample_serve_id)
    
    # 3. Visualize serve features
    visualizer.visualize_serve_features(sample_serve_id)
    
    # 4. Visualize joint trajectories
    visualizer.visualize_joint_trajectories(
        sample_serve_id, 
        joint_indices=[16, 15, 14, 13]  # Right arm + left wrist
    )
    
    # 5. Create serve animation
    visualizer.create_serve_animation(sample_serve_id, use_denormalized=True)
    
    # 6. Analyze overall dataset statistics
    visualizer.analyze_dataset_statistics()
    
    print("\nüéâ Visualization complete! All plots and animations saved.")
    print("\nüìã Available methods:")
    print("   - visualize_normalization_statistics()")
    print("   - compare_normalized_vs_denormalized(serve_id)")
    print("   - visualize_serve_features(serve_id)")  
    print("   - visualize_joint_trajectories(serve_id, joint_indices)")
    print("   - create_serve_animation(serve_id)")
    print("   - analyze_dataset_statistics()")
```
