```python
import pyvista as pv
import numpy as np

# --- 1. Load the VTK Mesh ---
# This loads the geometry (points) and topology (cells)
mesh = pv.read('/home/jtb3sud/G-PARC/demos/embryo/half_sph_mesh_061719.vtk')

# You can inspect the mesh to see what's inside
print("--- Mesh Info ---")
print(mesh)
print("\nAvailable data arrays in VTK:", mesh.array_names)

# --- 2. Load the TXT File Data ---
# This is a critical step and depends on your file's format.
# We'll use np.loadtxt(), which works well for simple columns of numbers.
# You may need to adjust this (e.g., use 'skiprows' or 'usecols').
try:
    # Load the text file
    custom_data = np.loadtxt('/home/jtb3sud/G-PARC/demos/embryo/half_sph_mesh_061719.txt')
    
    # --- 3. Combine the Data ---
    # We need to add the data from the .txt file to the mesh object.
    # We must ensure the data length matches the number of points or cells.
    
    if custom_data.size == mesh.n_points:
        # Data maps to points (nodes)
        mesh.point_data['my_custom_data'] = custom_data
        active_scalars_name = 'my_custom_data'
        print(f"\nSuccessfully added point data '{active_scalars_name}' to mesh.")
        
    elif custom_data.size == mesh.n_cells:
        # Data maps to cells (elements)
        mesh.cell_data['my_custom_data'] = custom_data
        active_scalars_name = 'my_custom_data'
        print(f"\nSuccessfully added cell data '{active_scalars_name}' to mesh.")
        
    else:
        print(f"\nWarning: Data size mismatch!")
        print(f"Mesh points: {mesh.n_points}, Mesh cells: {mesh.n_cells}")
        print(f"Data points in .txt: {custom_data.size}")
        active_scalars_name = None

except Exception as e:
    print(f"Could not load .txt file: {e}")
    active_scalars_name = None

# --- 4. Visualize the Mesh ---
# Create a PyVista plotter object
plotter = pv.Plotter(window_size=[800, 800])

# Add the mesh to the plotter
# 'scalars' tells it what data array to use for coloring
plotter.add_mesh(
    mesh, 
    scalars=active_scalars_name,  # Use the data we just added
    show_edges=True,             # Show the FEM mesh lines
    cmap='coolwarm',             # Color map
    scalar_bar_args={'title': 'Custom Data'} # Legend
)

# Set the camera position (optional, but helpful)
# 'xy' gives a good starting view for many 3D objects
plotter.camera_position = 'xy'

# Show the interactive plot window
print("Showing plot...")
plotter.show()

# You can also save a screenshot directly
# plotter.screenshot('my_fem_mesh.png')
```

```python
import pyvista as pv

# --- 1. Load the VTK Mesh ---
mesh = pv.read('/home/jtb3sud/G-PARC/demos/embryo/half_sph_mesh_061719.vtk')

# Inspect mesh (optional)
print(mesh)

# --- 2. Create an Interactive Plotter ---
plotter = pv.Plotter(window_size=[800, 800])

# Add the mesh to the scene
plotter.add_mesh(
    mesh,
    show_edges=True,
    color='lightblue',   # or choose any color you like
)

# Add a few interactive helpers
plotter.add_axes()
plotter.add_orientation_widget(pv.AxesActor())
plotter.add_text("Press X/Y/Z for views or rotate with mouse", font_size=10)

# --- 3. Choose Starting Camera Angle ---
plotter.view_xy()   # or .view_yz(), .view_xz()

# --- 4. Show Interactive Viewer ---
plotter.show()
```

```python
import pyvista as pv

# --- 1. Load the mesh ---
mesh = pv.read('/home/jtb3sud/G-PARC/demos/embryo/half_sph_mesh_061719.vtk')

# --- 2. Create an off-screen plotter (since you only need images) ---
plotter = pv.Plotter(off_screen=True, window_size=[800, 800])

# --- 3. Add the mesh ---
plotter.add_mesh(mesh, show_edges=True, color='lightblue')

# --- 4. Define camera positions ---
# Below the mesh (looking upward)
plotter.camera_position = [
    (0, 0, -1600),  # camera location below the mesh
    (0, 0, 0),     # focal point (center of mesh)
    (0, 1, 0)      # 'up' direction of the camera
]

# --- 5. Render and save ---
plotter.show(screenshot='half_sphere_bottom.png')
print("Saved view from underneath as 'half_sphere_bottom.png'")
```
